## Ploting

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(1, 10, 0.1)

y1 = x
ylog = np.log(x)
y2 = x**2
yexp = 2**x

plt.plot(x, y1, label='linear')
plt.plot(x, ylog, label='logrithm')
plt.plot(x, y2, label='power')
plt.plot(x, yexp, label='exponential')

plt.grid(True)
plt.axis([0, 10, 0, 100])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Order of growth')

plt.legend(loc='upper center')

plt.savefig('./IMG/order_of_growth.png',dpi=600)
plt.show()
```

## Examples

### pyplot 2D

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import numpy as np
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='b')

plt.xlim(0, 2*np.pi)
plt.ylim(-1.5, 1.5)

plt.title("y = Sin(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(True, linestyle='--')

patch = mpatches.Patch(label='Legend:')
red_line = mlines.Line2D([], [], color='red', label='y = Sin(x)')
blue_line = mlines.Line2D([], [], color='blue', label='y = Cos(x)')
plt.legend(handles=[patch, red_line, blue_line])

plt.savefig('data/sincos.png', dpi=600)
```
